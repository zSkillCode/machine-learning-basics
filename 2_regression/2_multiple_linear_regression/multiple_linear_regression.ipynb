<h1>Importing the libraries</h1>

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

<h1>Importing the dataset</h1>

```
dataset = pd.read_csv('data.csv')
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

<h1>Encoding categorial data</h1>

```
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

column_transformer = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])],
                                       remainder='passthrough')
X = np.array(column_transformer.fit_transform(X))
```

<h1>Splitting the dataset into the Training set and Test set</h1>

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

<h1>Predicting the Test set results</h1>

```
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

<h1>Predicting the Test set results</h1>

```
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.column_stack((y_pred, y_test)))
```
