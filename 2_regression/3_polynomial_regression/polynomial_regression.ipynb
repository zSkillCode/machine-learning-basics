<h1>Importing the libraries</h1>

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

<h1>Importing the dataset</h1>

```
dataset = pd.read_csv('data.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

<h1>Training the Linear Regression model on the whole dataset</h1>

```
from sklearn.linear_model import LinearRegression

linear_regressor = LinearRegression()
linear_regressor.fit(X, y)
```

<h1>Training the Polynomial Regression model on the whole dataset</h1>

```
from sklearn.preprocessing import PolynomialFeatures

polynomial_regressor = PolynomialFeatures(degree=4)
X_poly = polynomial_regressor.fit_transform(X, y)
linear_regressor_2 = LinearRegression()
linear_regressor_2.fit(X_poly, y)
```

<h1>Visualising the Linear Regression results</h1>

```
plt.scatter(X, y, color='red')
plt.plot(X, linear_regressor.predict(X), color='blue')
plt.title('Truth or Bluff (Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

<h1>Visualising the Polynomial Regression results</h1>

```
plt.scatter(X, y, color='red')
plt.plot(X, linear_regressor_2.predict(X_poly), color='blue')
plt.title('Truth or Bluff (Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()
```

<h1>Predicting a new result with Linear Regression</h1>

```
linear_regressor.predict([[6.5]])
```

<h1>Predicting a new result with Polynomial Regression</h1>

```
linear_regressor_2.predict(polynomial_regressor.fit_transform([[6.5]]))
```
