<h1>Importing the libraries</h1>

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

<h1>Importing the dataset</h1>

```python
dataset = pd.read_csv('data.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

<h1>Splitting the dataset into the Training set and Test set</h1>

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=543)
```

<h1>Training the Simple Linear Regression model on the Training set</h1>

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

<h1>Predicting the Test set results</h1>

```python
y_pred = regressor.predict(X_test)
```

```python
y_pred
```

<h1>Visualising the Training set results</h1>

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, regressor.predict(X_train), color='blue')
plt.title('Salary vs Experience (Training Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

<h1>Visualising the Test set results</h1>

```python
plt.scatter(X_test, y_test, color='red')
plt.plot(X_test, y_pred, color='blue')
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
